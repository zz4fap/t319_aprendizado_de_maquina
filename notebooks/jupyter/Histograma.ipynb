```python
# Importando as bibliotecas necessárias.
import numpy as np
import matplotlib.pyplot as plt

"""
Com a função mágica abaixo, a saída dos comandos de plot do biblioteca 
matplotlib é exibida inline, ou seja, diretamente abaixo da célula de 
código que a produziu e de forma estática. Os plots resultantes também 
serão armazenadas no notebook. Para mais informações: 
https://ipython.readthedocs.io/en/stable/interactive/tutorial.html#magics-explained
"""
%matplotlib inline
```

```python
# Reseta o gerador de números pseudo-aleatórios (PN).
seed = 42
np.random.seed(seed)
```

### Gerandos dados.

```python
# Número de exemplos.
N = 1000000

# Criando um vetor coluna x, com dimensão Nx1 e 
# valores retirados de uma distribuição uniforme 
# no intervalo de 0 a 1.
x = np.random.rand(N,1)

# Criando um vetor coluna w, com dimensão Nx1 e 
# valores retirados de uma distribuição normal
# com média 0 e variância 0.01.
w = np.sqrt(0.01)*np.random.randn(N,1)

# Função original ou verdadeira.
y = 1 + 2*x 

# Versão ruidosa de y.
y_noisy = y + w
```

### Plotando as funções.

```python
plt.figure()
plt.plot(x, y_noisy, '.b', label='Função ruidosa')
plt.plot(x, y, 'k', label='Função original')
plt.xlabel('x', fontsize=14)
plt.ylabel('y', fontsize=14)
plt.legend()
plt.grid()
plt.show()
```

### Plotando a PDF e CDF do sinal ruidoso.

```python
# histograma (pdf)
plt.subplot(1, 2, 1)
plt.title('PDF')
plt.hist(y_noisy, bins=100, density=True, color='b')

# CDF empírica
plt.subplot(1, 2, 2)
plt.title('CDF')
plt.hist(y_noisy, bins=100, density=True, color='g', cumulative=True)

# salva figura em arquivo
plt.savefig('histogram.png')
```
