```python
# importing libraries
import numpy as np
import random
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

%matplotlib notebook
```

```python
def f(x):
    return 2*(x**2)
```

```python
N = 1000
x = np.linspace(-2,2,N)

c = -2
y0 = c + f(x)

c = -1
y1 = c + f(x)

c = 0
y2 = c + f(x)

c = 1
y3 = c + f(x)

c = 2
y4 = c + f(x)

plt.figure()
plt.plot(x,y0, label='c=-2')
plt.plot(x,y1, label='c=-1')
plt.plot(x,y2, label='c=0')
plt.plot(x,y3, label='c=+1')
plt.plot(x,y4, label='c=+2')
plt.grid()
plt.legend()
plt.show()
```

```python
N = 200
x = np.linspace(-2,2,N)
y = np.linspace(-2,2,N)

X, Y = np.meshgrid(x, y)

print(f(X).shape)
print(Y.shape)

# Generate points for plotting the cost-function surface.
C = np.zeros((N,N))
for iter1 in range(0, N):
    for iter2 in range(0, N):
        C[iter1][iter2] = Y[iter1][iter2] - f(X[iter1][iter2])

# Plot cost-function surface.
fig = plt.figure()
ax = fig.gca(projection='3d')
surf = ax.plot_surface(X, Y, C, cmap=cm.coolwarm, linewidth=0, antialiased=False)
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_zlabel('$c$')
#ax.view_init(20, 45)
#Show the plot.
fig.show()
```

```python
N = 200
x = np.linspace(-2,2,N)
c = np.linspace(-2,2,N)

X, C = np.meshgrid(x, c)

# Generate points for plotting the cost-function surface.
Y = np.zeros((N,N))
for iter1 in range(0, N):
    for iter2 in range(0, N):
        Y[iter1][iter2] = C[iter1][iter2] + f(X[iter1][iter2])

# Plot cost-function surface.
fig = plt.figure()
ax = fig.gca(projection='3d')
surf = ax.plot_surface(X, Y, C, cmap=cm.coolwarm, linewidth=0, antialiased=False)
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_zlabel('$c$')
#ax.view_init(20, 45)
#Show the plot.
fig.show()
```
