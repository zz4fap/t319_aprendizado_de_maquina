```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Always reset the pseudo-random numbers generator to a known value so that your results are always the same.
np.random.seed(1234)
```

```python
# Definindo número de exemplos.
N = 1000

# Atributo.
x = np.random.rand(N,1)

# Ruído.
w = np.sqrt(0.1)*np.random.randn(N,1)

# Função objetivo.
y = 1 + x

# Função observável (ruidosa).
y_noisy = y + w
```

```python
plt.plot(x,y_noisy,'.', label='Função ruidosa')
plt.plot(x,y, 'k', label='Função objetivo', linewidth=5)
plt.xlabel('$x_1$')
plt.ylabel('$y$')
plt.grid()
plt.legend()
plt.savefig("observable_function.png", dpi=600)
plt.show()
```

```python
# Create attribute matrix.
X = np.c_[np.ones((N, 1)), x]

# Equação normal.
a_opt = np.linalg.pinv((X.T).dot(X)).dot((X.T).dot(y_noisy))

# Prediction.
yhat = X.dot(a_opt)

# MSE calculation.
Joptimum = (1.0/N)*np.sum(np.power((y_noisy - yhat), 2))

print('Joptimum:', Joptimum)
```

```python
# Plot pair example/label.
fig = plt.figure()
plt.plot(x, y_noisy, '.', label='Observable function')
plt.plot(x, y, 'k', label='True function', linewidth=7)
plt.plot(x, yhat, 'r--', label='Hypothesis function')
plt.xlabel('x',fontsize=14)
plt.ylabel('y',fontsize=14)
plt.legend()
plt.grid()
plt.savefig('regression.png', dpi=600)
plt.show()
```
