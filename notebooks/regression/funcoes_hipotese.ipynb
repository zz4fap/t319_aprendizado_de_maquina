```python
import numpy as np
import matplotlib.pyplot as plt
```

# Reta

```python
x1 = np.linspace(-10,10,100)

y = 1.2 + 2.4*x1
```

```python
plt.figure()
plt.plot(x1,y)
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.savefig('linha_reta.png', dpi=600)
plt.show()
```

# Superfície 2D

```python
x1 = np.linspace(-10,10,100)
x2 = np.linspace(-10,10,100)

y = 1.2 + 2.4*x1 + 3.3*x2
```

```python
# para figuras interativas usar 'notebook' ao inves de 'inline'
%matplotlib notebook

X1, X2 = np.meshgrid(x1, x2)
Y = 1.2 + 2.4*X1 + 3.3*X2

ax = plt.subplot(111, projection='3d')
ax.plot_surface(X1, X2, Y)
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('y')
plt.savefig('superficie.png', dpi=600)
plt.show()
```
